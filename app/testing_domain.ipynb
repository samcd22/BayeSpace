```python
from visualisation_toolbox.domain import Domain

domain = Domain(n_dims=1, domain_select='log').add_domain_param('min', 0).add_domain_param('max', 1).add_domain_param('n_points', 10)
points = domain.build_domain()
display(points)
domain.plot_domain()
```

```python
domain = Domain(n_dims = 2, domain_select = 'rectangular')
domain = domain.add_domain_param('min_x', 0).add_domain_param('max_x', 1).add_domain_param('n_points_x', 10)
domain = domain.add_domain_param('min_y', 0).add_domain_param('max_y', 1).add_domain_param('n_points_y', 10)

points = domain.build_domain()
# points = domain.cut_domain([0.5,0.5], [-1,1])
points = domain.transform_domain([[0,1],[2,1]])
# points = domain.cut_domain([0.5,0.5], [0,-1])
display(points)
domain.plot_domain()
```

```python
import numpy as np
domain = Domain(n_dims = 2, domain_select = 'circular')
domain = domain.add_domain_param('radius', 10).add_domain_param('mass', 1000).add_domain_param('center', [0,8])

points = domain.build_domain()
points = domain.cut_domain([0,0], [0,-1])

display(points)
domain.plot_domain()
```

```python
domain = Domain(n_dims = 2, domain_select = 'triangular')
domain = domain.add_domain_param('coord1', [0,0]).add_domain_param('coord2', [1,2]).add_domain_param('coord3', [2,1]).add_domain_param('mass', 1000)
points = domain.build_domain()
display(points)
domain.plot_domain()
```

```python
domain = Domain(n_dims=2, domain_select='polygonal')
domain.add_domain_param('vertices', [[0, 0], [4, 1], [3, 4], [1, 5], [-1, 2]])
domain.add_domain_param('mass', 500)  # Density-based parameter

points = domain.build_domain()
print(len(points))  # Number of points should match density
domain.plot_domain()  # Plot the domain
```

```python
domain = Domain(n_dims=3, domain_select='cuboidal')
domain.add_domain_param('min_x', 0).add_domain_param('max_x', 1).add_domain_param('n_points_x', 10)
domain.add_domain_param('min_y', 0).add_domain_param('max_y', 1).add_domain_param('n_points_y', 10)
domain.add_domain_param('min_z', 0).add_domain_param('max_z', 1).add_domain_param('n_points_z', 10)

points = domain.build_domain()
cross_section_points, projected_cross_section_points = domain.apply_cross_section([0.4,0.4,0.4], [0,1,0], 0.1)
domain.plot_cross_sections()


# print(len(points))  # Number of points should match density
domain.plot_domain()  # Plot the domain
```

```python
domain = Domain(n_dims=3, domain_select='cylindrical')
domain.add_domain_param('radius', 10).add_domain_param('height', 10).add_domain_param('angle', [0,np.pi/2]).add_domain_param('mass', 10000)

points = domain.build_domain()

points = domain.transform_domain([[1,0,0],[0, 0,1],[0, 1,0]])

print(len(points))  # Number of points should match density
domain.plot_domain()  # Plot the domain
```

```python
domain = Domain(n_dims=3, domain_select='spherical')
domain.add_domain_param('radius', 1000)\
        .add_domain_param('mass', 100000)\
        .add_domain_param('theta', [-np.pi/4,np.pi/4])\
        .add_domain_param('phi', [-np.pi/4,np.pi/4])\
        .add_domain_param('center', [0,0,5])

points = domain.build_domain()
points = domain.cut_domain([0,0,0], [0,0,1])

domain.apply_cross_section([0,0,5], [0,0,1], 10)
domain.apply_cross_section([0,0,0], [0,1,0], 10)

domain.plot_domain()
domain.plot_cross_sections()
```
