```python
import numpy as np
from matplotlib import pyplot as plt

from inference_toolbox.domain import Domain

domain = Domain('cone_from_source_z_limited', resolution=100)
domain.add_domain_param('r', 1000)
domain.add_domain_param('theta', np.pi/8)
domain.add_domain_param('source', [0,0,10])

points = domain.create_domain()

fig = plt.figure()

# Defines the lower bound subplot
ax1 = fig.add_subplot(111, projection = '3d')
ax1.scatter(points[:,0],points[:,1],points[:,2], s=0.1)
plt.show()
```
